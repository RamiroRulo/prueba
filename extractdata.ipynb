```python
def mm(df,t,veces1,veces2,valor):
    global m
    df=df
    valor1=df[valor].rolling(veces1).mean()[t]
    valor2=df[valor].rolling(veces2).mean()[t]
    
    
    m=valor1/valor2
    return m

def stdm(df, t, veces1,veces2,valor):
    global s
    df=df
    valor1=df[valor].rolling(veces1).std()[t]
    valor2=df[valor].rolling(veces2).std()[t]
    
    
    s=valor1/valor2
    return s
def relacion_index(df, dfi,t, veces1, veces2, valor):
    global r
    valorA1=dfi[valor].dropna().rolling(veces1).mean()[t]
    valorB1=df[valor].dropna().rolling(veces1).mean()[t]
    valorA2=dfi[valor].dropna().rolling(veces2).mean()[t]
    valorB2=df[valor].dropna().rolling(veces2).mean()[t]
    
    r=(valorA1/valorB1)/(valorA2/valorB2)
    return r
   
    
    
  
```
